Task 1 - Sum of Nested Lists:

```python
def sum_nested_list(nested_list):

  total=0;
  for x in nested_list:
    if isinstance(x,list):
      total+=sum_nested_list(x);
    else:
      total+=x;
  return total;
print (sum_nested_list( [1, [2, [3, 4], 5], 6, [7, 8]]));
```

Task 2 - Generate All Permutations of a String:

```python
from itertools import permutations
def generate_permutations(s):
  for x in permutations(s):
    print(''.join(x));

generate_permutations("abc");
```

Task 3 - Directory Size Calculation:

```python

def calculate_directory_size(directory_structure):
   total=0;
   for x in directory_structure:
     if isinstance(directory_structure[x],dict):
       total+=calculate_directory_size(directory_structure[x]);
     else:
       total+=directory_structure[x];
   return total;
directory_structure = {
"file1.txt": 200,
"file2.txt": 300,
"subdir1": {
"file3.txt": 400,
"file4.txt": 100
},
"subdir2": {
"subsubdir1": {
"file5.txt": 250
},
"file6.txt": 150
}
}

print(calculate_directory_size(directory_structure));
```

8.2 Dynamic Programming:

Task 1 - Coin Change Problem:

```python
def min_coins(coins, amount):

  dp = [float("inf")] * (amount + 1)
  dp[0] = 0;
  for coin in coins:
    for i in range(coin, amount + 1):
      dp[i] = min(dp[i], dp[i - coin] + 1)
  return dp[amount] if dp[amount] != float("inf") else -1
print(min_coins([1,2,5],11));
```

Task 2 - Longest Common Subsequence (LCS):

```python
def longest_common_subsequence(a,b):
    dp=[[0]*(len(b)+1) for _ in range(len(a)+1)]
    for i in range(1,len(a)+1):
        for j in range(1,len(b)+1):
            if a[i-1]==b[j-1]:
                dp[i][j]=dp[i-1][j-1]+1
            else:
                dp[i][j]=max(dp[i-1][j], dp[i][j-1])
    return dp[-1][-1]

print(longest_common_subsequence("abcde","ace"))
```

Task 3 - 0/1 Knapsack Problem:



```python
def knapsack(weights, values, cap):
    n=len(weights)
    dp=[[0]*(cap+1) for _ in range(n+1)]
    for i in range(1,n+1):
        for c in range(cap+1):
            if weights[i-1] <= c:
                dp[i][c]=max(dp[i-1][c],dp[i-1][c-weights[i-1]]+values[i-1])
            else:
                dp[i][c]=dp[i-1][c]
    return dp[n][cap]

print(knapsack([1,3,4,5],[1,4,5,7],7))
```
